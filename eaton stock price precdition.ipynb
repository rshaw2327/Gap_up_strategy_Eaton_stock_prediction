```python
import pandas as pd

def bull_analysis(df, gap_up=0.25, num_days=30):
    """
    Analyze 'gap-up' events in price data.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame with columns: 'Date', 'Open', 'High', 'Close', 'Returns'
    gap_up : float
        Minimum % return to define a 'gap-up' day (e.g., 0.35 = 0.35%)
    num_days : int
        Number of days to track performance after a gap-up

    Returns
    -------
    trades_df : pd.DataFrame
        Table of each trade with start date, end date, prices, and returns
    """

    trades = []  # list to hold results
    flag = False
    counter = 0
    start_date = None
    start_price = 0
    highest_price = 0

    for i in range(len(df)):
        if not flag:  # look for a gap-up trigger
            if df.loc[i, 'Returns'] >= gap_up:
                flag = True
                counter = 1
                start_date = df.loc[i, 'Date']
                start_price = df.loc[i, 'Open']
                highest_price = df.loc[i, 'High']
        else:
            counter += 1
            # update the high price within the observation window
            if df.loc[i, 'High'] > highest_price:
                highest_price = df.loc[i, 'High']

            if counter == num_days:
                end_date = df.loc[i, 'Date']
                returns_pct = round((highest_price - start_price) / start_price * 100, 2)
                trades.append({
                    'Start_Date': start_date,
                    'Start_Price': start_price,
                    'End_Date': end_date,
                    'Highest_Price': highest_price,
                    'Return_%': returns_pct
                })
                # reset for next potential trade
                flag = False
                counter = 0
                start_date = None
                start_price = 0
                highest_price = 0

    # convert to DataFrame
    trades_df = pd.DataFrame(trades)

    if trades_df.empty:
        print("‚ö†Ô∏è No gap-up events found above threshold.")
        return trades_df

    # summary statistics
    avg_return = trades_df['Return_%'].mean()
    max_return = trades_df['Return_%'].max()
    min_return = trades_df['Return_%'].min()

    print(f"‚úÖ Total gap-up trades found: {len(trades_df)}")
    print(f"üìà Average Return: {avg_return:.2f}%")
    print(f"üîù Max Return: {max_return:.2f}%")
    print(f"üîª Min Return: {min_return:.2f}%")

    return trades_df
```
